```python
x = randn(10000) # create a random number vector with 10000 variables. 
plot(x) #plot random numbers
grid(True) #use grid in the plot

#add labels to the chart
title('Sample Values') 
xlabel('Sample')
ylabel('Value')
```

```python
h = hist(x,100) # make a histogram of returns, using 100 bins. 
#h is used as a return variable to avoid text output

print('Mean: %.2f' % x.mean()) # print mean 
print('Standard deviation: %.2f' % x.std()) # print sigma, note the 2 decimal precision notation
```
